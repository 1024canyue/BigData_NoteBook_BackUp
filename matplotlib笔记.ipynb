# 给图表添加标注

```python
x=np.linspace(-5,5,50)    #每个数据点x坐标
y=2*x+1    #每个数据点y坐标
```

```python
x0 = 1   #要强调的那个点
y0 =2 * x0 + 1
```

```python
plt.plot(x,y)  #绘制图形

#下面这些是画常用坐标轴的，可复制直接用
ax = plt.gca()
ax.spines['right'].set_color('none')#隐藏右边黑框
ax.spines['top'].set_color('none')   #隐藏上边黑框
ax.xaxis.set_ticks_position('bottom')   #x轴坐标刻度显示在底部
ax.spines['bottom'].set_position(('data', 0))   #x轴居中
ax.yaxis.set_ticks_position('left')    #y轴坐标刻度显示在左侧
ax.spines['left'].set_position(('data', 0))   #y轴居中

plt.scatter(x0,y0,s=50,color="b")#绘制那个点   大小50，颜色蓝色

#绘制那条指示线（两点坐标连线，样式k--即黑色，长长虚线,线宽2.5）
plt.plot([x0,x0],[y0,0],"k--",lw=2.5)   

# 右边标记
plt.annotate(r'$2x+1=%s$'%y0  #文字
             ,xy=(x0,y0)  #坐标
             ,xytext=(+30,-30),textcoords="offset points"  #偏移
             ,fontsize=16  #字体大小
            ,arrowprops=dict(arrowstyle="->",connectionstyle="arc3,rad=.2"))  #箭头以及箭头样式
#左边文本标记
plt.text(-6,6  #显示的位置
         ,r"$This\ is\ the\ some\ text.\ \mu\ \sigma_i\ \alpha_t$" #花里胡哨的各种文字和符号(μ 西塔 阿尔法)
        ,fontdict={"size":11,"color":"r"})  #字体样式

plt.show()
```

### 饼 图
pie(x, explode=None, labels=None, colors=None, autopct=None,
        pctdistance=0.6, shadow=False, labeldistance=1.1, startangle=None,
        radius=None, counterclock=True, wedgeprops=None, textprops=None,
        center=(0, 0), frame=False, rotatelabels=False, hold=None, data=None)
- x       :(每一块)的比例，如果sum(x) > 1会使用sum(x)归一化；
- labels  :(每一块)饼图外侧显示的说明文字；
- explode :(每一块)离开中心距离；
- startangle :起始绘制角度,默认图是从x轴正方向逆时针画起,如设定=90则从y轴正方向画起；
- shadow  :在饼图下面画一个阴影。默认值：False，即不画阴影；
- labeldistance :label标记的绘制位置,相对于半径的比例，默认值为1.1, 如<1则绘制在饼图内侧；
- autopct :控制饼图内百分比设置,可以使用format字符串或者format function '%1.1f'指小数点前后位数(没有用空格补齐)；
- pctdistance :类似于labeldistance,指定autopct的位置刻度,默认值为0.6；
- radius  :控制饼图半径，默认值为1；
- counterclock ：指定指针方向；布尔值，可选参数，默认为：True，即逆时针。将值改为False即可改为顺时针。
- wedgeprops ：字典类型，可选参数，默认值：None。参数字典传递给wedge对象用来画一个饼图。例如：wedgeprops={'linewidth':3}设置wedge线宽为3。
- textprops ：设置标签（labels）和比例文字的格式；字典类型，可选参数，默认值为：None。传递给text对象的字典参数。
- center ：浮点类型的列表，可选参数，默认值：(0,0)。图标中心位置。
- frame ：布尔类型，可选参数，默认值：False。如果是true，绘制带有表的轴框架。
- rotatelabels ：布尔类型，可选参数，默认为：False。如果为True，旋转每个label到指定的角度。<br>

# 3D图表

```python
import numpy as np
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D
```

```python
fig = plt.figure()
ax = Axes3D(fig)   #添加3D坐标轴
# XY数据
X = np.arange(-4, 4, 0.25)
Y = np.arange(-4, 4, 0.25)
X, Y = np.meshgrid(X, Y)   #生成网格点坐标矩阵
R = np.sqrt(X ** 2 + Y ** 2)
# 高度值Z
Z = np.sin(R)

#3D模型
ax.plot_surface(X, Y, Z  #三个坐标轴
                , rstride=2  #跨度(竖向)（降低可提高模型的精细程度）
                , cstride=4 #跨度(横向)（降低可提高模型的精细程度）
                , edgecolor="black"   #黑色的网格线
                , cmap=plt.get_cmap('rainbow'))   #颜色（彩虹）

#等高线
ax.contourf(X, Y, Z
            , zdir='z'  #从那个方向看
            , offset=-2   #把图放在这个轴对应的值的位置上
            , cmap=plt.get_cmap('rainbow'))

ax.set_zlim(-2, 2)  #显示z轴-2~2的区域

plt.show()   #jupyter不能转，如果用它自带的窗口是可以转动的
```

# 一个窗口 figure画多个图

### 方法1

```python
plt.figure()
plt.subplot(2,1,1)  #两行一列样式，第一个位置
plt.plot([0,1],[0,1])   #0,0到1,1连线

plt.subplot(2,3,4)  #两行三列样式，第四个位置 （1-3位置被占用了，原来的会被覆盖）
plt.plot([1,0],[0,1])   #0,0到1,1连线

plt.subplot(2,3,5)  #两行三列，第五个位置
plt.plot([1,0],[0,1])   #0,0到1,1连线

plt.subplot(2,3,6)  #两行三列，第六个位置
plt.plot([1,0],[1,0])   #0,0到1,1连线

plt.show()
```

### 方法2

```python
import matplotlib.gridspec as gridspec   #需要额外导入

plt.figure()

ax1 = plt.subplot2grid((3, 3)      #三行三列
                       , (0, 0)    #0行0列位置    （位置都是从0开始算）
                       , colspan=3  #行跨度为3（横着占三个空间）
                       #,rowspan=1   #列跨度为1（竖着占一个空间）
                      )  
ax1.plot([1, 2], [1, 2])
ax1.set_title('ax1_title')

ax2 = plt.subplot2grid((3, 3), (1, 0), colspan=2)   #三行三列，1行0列位置，横着占2个位置

ax3 = plt.subplot2grid((3, 3), (1, 2), rowspan=2)   #三行三列，1行2列位置，竖着占2个位置

ax4 = plt.subplot2grid((3, 3), (2, 0))      #三行三列，2行0列位置，占1个位置
ax4.scatter([1, 2], [2, 2])
ax4.set_xlabel('ax4_x')
ax4.set_ylabel('ax4_y')

ax5 = plt.subplot2grid((3, 3), (2, 1))   #三行三列，2行1列位置，占1个位置

plt.show()
```

### 方法3：图中图

```python
fig = plt.figure()    #把plt.figure()定义成fig
x = [1, 2, 3, 4, 5, 6, 7]
y = [1, 3, 4, 2, 5, 8, 6]

           #add_axes新增子区域
ax1 = fig.add_axes([0.1   #相对figure左右间距10%
                    , 0.1 #相对figure上下间距10%
                    , 0.8 #宽是figure的80%
                    , 0.8])  #高是figure的80%
ax1.plot(x, y, 'r')
ax1.set_xlabel('x')
ax1.set_ylabel('y')
ax1.set_title('title')

#下同，再添加一张
ax2 = fig.add_axes([0.2, 0.6, 0.25, 0.25])
ax2.plot(y, x, 'b')
ax2.set_xlabel('x')
ax2.set_ylabel('y')
ax2.set_title('title inside 1')

plt.show()
```

# 主/次坐标轴

```python
x = np.arange(0, 10, 0.1)
y1 = 0.05 * x**2
y2 = -1 *y1

fig, ax1 = plt.subplots()   #创建总画布，并接受返回的figure与axes（轴线）

ax2 = ax1.twinx()   #ax2（右边的那条y轴加x轴）是ax1（左边那条y轴加x轴）的镜像，即把左边的y轴复制到右边
ax1.plot(x, y1, 'g-')   #g- 绿色 实线
ax2.plot(x, y2, 'b-')   #b- 蓝色 实线

ax1.set_xlabel('X data')    #设置这三个轴的标题
ax1.set_ylabel('Y1 data', color='g')
ax2.set_ylabel('Y2 data', color='b')

plt.show()
```

两个图型共享一个x轴，所以x轴是主轴，它们都有自己独立的y轴，所以y轴是次轴

# Animation动画
pupyter没配置好不能显示动画效果，若需演示可以请复制到其他地方运行

```python
from matplotlib import animation   #需另外导入动画模块

fig, ax = plt.subplots()  #创建总画布，并接受返回的figure与axes（轴线）

x = np.arange(0, 2*np.pi, 0.01)   #数据
line, = ax.plot(x, np.sin(x))   #sin函数的图像


def animate(i):   #传入某一帧 0-frames参数
    line.set_ydata(np.sin(x + i/10.0))  #更新y数据
    return line,  #返回这时sin函数的图像


def init():   #原来输出起始图像
    line.set_ydata(np.sin(x))
    return line,  #输出起始图像

ani = animation.FuncAnimation(
                                fig=fig   #把刚刚创建总画布是返回的figure传入
                                , func=animate   #动画
                                , frames=63   #总帧数
                                , init_func=init  #起始的图像
                                , interval=15  #间隔多少毫秒更新一次动画
                                , blit=False  #更新整张图T/只更新变化的数据F（MAC用户只能用F）
)

plt.show()
```
