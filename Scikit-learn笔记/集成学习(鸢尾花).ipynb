```python
from sklearn import datasets  #加载鸢尾花数据集
iris=datasets.load_iris()
```

```python
X = iris.data   #拆分特征与标签
Y = iris.target
```

```python
from sklearn.model_selection import train_test_split   #拆分训练与测试集
X_train,X_test,Y_train,Y_test = train_test_split(X,Y,test_size=0.2)
```

```python
from sklearn.tree import DecisionTreeClassifier #导入决策树分类器
from sklearn.svm import SVC  #导入支持向量机分类器
from sklearn.ensemble import RandomForestClassifier  #随机森林（其实这本事就是一个集成学习）

dtree_clf = DecisionTreeClassifier(random_state=42)   #实例化这三个分类器
svm_clf = SVC(random_state=43)
rf_clf = RandomForestClassifier(random_state=44,n_jobs=-1)  #,n_jobs=-1使用所有的CPU(会快一点)

from sklearn.ensemble import VotingClassifier  #投票分类器  (用来集成上面的分类器)
voting_clf = VotingClassifier(
    estimators=[("0",dtree_clf),("1",svm_clf),("2",rf_clf)],  #为演示前面字符串可以随意填，但不建议这么做
    voting="hard"  #硬投票
    #voting = 'hard'：根据少数服从多数来定最终结果（硬投票）
    #voting = 'soft':准确率最高的模型为最终的预测结果（软投票）
)
voting_clf.fit(X_train,Y_train)
```

```python
from sklearn.model_selection import cross_val_score
score=cross_val_score(voting_clf,X,Y   #交叉验证评分
                ,cv=5     #分成五组
                ,scoring="accuracy"     #判断准确度
               ).mean()   #取平均值
score
```

```python
import pandas as pd
feature_names=["花萼长度(cm)","花萼宽度(cm)","花瓣长度(cm)","花瓣宽度(cm)"]
target_names=["山鸢尾","变色鸢尾","维吉尼亚鸢尾"]
predict,test=[],[]

for i in range(len(voting_clf.predict(X_test))):  #预测结果转文字
    predict.append(target_names[voting_clf.predict(X_test)[i]])
    
for i in range(len(Y_test)):     #实际结果转文字
    test.append(target_names[Y_test[i]])

pd.concat([pd.DataFrame(X_test,columns=feature_names),
           pd.DataFrame(predict,columns=["预测分类"]),
           pd.DataFrame(test,columns=["实际分类"])],
           axis=1)
```

```python
print("准确度：{:.2f}%".format(score*100))
```
