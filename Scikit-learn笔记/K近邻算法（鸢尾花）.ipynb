```python
import numpy as np
import pandas as pd
from sklearn import datasets  #调用sklearn自带数据集
from sklearn.model_selection import train_test_split #用来随机划分样本数据为训练集和测试集 #fsklearn.cross_validation包已弃用
from sklearn.neighbors import KNeighborsClassifier  #实现K近邻算法的一个类
```

```python
iris = datasets.load_iris()   #鸢尾花数据集
```

```python
iris_x =iris.data   #属性  （花萼长度、花萼宽度、花瓣长度、花瓣宽度）
iris_x[:3]
```

```python
iris_y =iris.target  #分类  （0：山鸢尾、1：变色鸢尾，2：维吉尼亚鸢尾）
iris_y
```

```python
#区分训练集与测试集，测试集比例为20%
X_train,X_test,Y_train,Y_test=train_test_split(iris_x,iris_y,test_size=0.2)
```

```python
knn = KNeighborsClassifier()  #将K近邻算法定义成knn，注意括号
clf=knn.fit(X_train,Y_train)
```

```python
print("预测结果:",clf.predict(X_test))
print("实际结果:",Y_test)
print("准确率：",clf.score(X_test,Y_test)*100,"%")
```

```python
predict,test=[],[]
class_names=["山鸢尾","变色鸢尾","维吉尼亚鸢尾"]

for i in range(len(clf.predict(X_test))):    #预测结果转文字
    predict.append(class_names[clf.predict(X_test)[i]])
    
for i in range(len(Y_test)):     #实际结果转文字
    test.append(class_names[Y_test[i]])
    
pd.concat([pd.DataFrame(X_test,columns=["花萼长度","花萼宽度","花瓣长度","花瓣宽度"])   #测试集数据
          ,pd.DataFrame(predict,columns=["预测分类"])     #预测结果（文字）
          ,pd.DataFrame(test,columns=["实际分类"])]   #实际结果（文字）
          ,axis=1)   #扩充行
```
