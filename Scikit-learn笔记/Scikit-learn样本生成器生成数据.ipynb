## 1:聚类样本生成器make_blobs()
主要是用来创建聚类的数据集，对每个聚类的中心，标准差都有很好的控制。
* 参考：https://www.cnblogs.com/shanger/p/12068518.html

```python
from  sklearn.datasets import make_blobs
import matplotlib.pyplot as plt
```

```python
# 创建仿真聚类数据集
X1, Y1 = make_blobs(
                    n_samples=1000,   # 待生成的样本的总数
                    n_features=2,    # 每个样本的特征数
                    centers=3,       # 要生成的样本中心（类别）数，或者是确定的中心点
                    cluster_std=1.0,     # 数据方差，代表簇的聚合程度
                    center_box=(-10.0, 10.0),   #中心确定之后的数据边界，亦即每个簇的上下限
                    shuffle=True,        # 是否将样本打乱
                    random_state=666)    #随机生成器的种子
plt.scatter(X1[:, 0], X1[:, 1],c=Y1,s=6)
plt.show()
```

```python
print(X1[:5])  #X是样本的特征属性
print(Y1[:20])   #y是样本的分类标签
```

## 2:分类样本集生成器 make_classification（）
* 参考：https://blog.csdn.net/genghaihua/article/details/81154566

```python
from sklearn.datasets import make_classification
```

```python
X2, Y2 = make_classification(n_samples=1000,   #样本个数
                             n_features=2,     #特征个数
                             n_redundant=0,    #冗余特征数
                             n_classes=2,      #输出的标签数量或类别(默认=2)
                             n_informative=1,  #多信息特征个数
                             n_clusters_per_class=1)  #每个class中cluster数量
plt.scatter(X2[:, 0], X2[:, 1],c=Y2,s=6)
plt.show()
```

```python
print(X2[:5])  #X是样本的特征属性
print(Y2[:10])   #Y是样本的分类标签
```

## 3：分组正态分布混合样本生成器 
## make_gaussian-quantiles()

```python
from sklearn.datasets import make_gaussian_quantiles
```

```python
X3,Y3=make_gaussian_quantiles(n_samples=1000,  #生成样本个数
                              n_features=2,    #正态分布维数
                              n_classes=3      #数据在正态分布中按分位数分配的组数
                              ,mean=[1,2],   #特征均值
                              cov=2)   #样本谐方差系数

plt.scatter(X3[:,0],X3[:,1],c=Y3,s=6)
plt.show()
```

```python
print(X3[:5])  #X是样本的特征属性
print(Y3[:10])   #Y是样本的分类标签
```

## 4：回归模型随机数据生成器 make_regression（）

```python
from sklearn.datasets import make_regression
```

```python
X4,Y4,coef=make_regression(  #返回样本特征X，样本分类Y，回归系数coef
                    n_samples=1000,   #生成样本数量
                    n_features=1,     #样本特征数
                    noise=13,         #噪声数据
                    coef=True)        #是否返回随机系数
plt.scatter(X4,Y4,c='b',s=3)
plt.plot(X4,X4*coef,c='r')  #输出回归系数
plt.show()
```

```python
print(X4[:5])  #X是样本的特征属性
print(Y4[:10])   #Y是样本的分类标签
print("回归系数:",coef)
```
